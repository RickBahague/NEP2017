```python
from openpyxl import *
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.style.use('ggplot')
from pyspark.sql.functions import *
```

```python
#Amount is in Thousands PHP
data = pd.read_csv("data/nep2017.csv",sep="|")
denr_funds = data[data.UACS_DPT_DSC=='Department of Environment and Natural Resources (DENR)']
```

```python
denr_funds.head(2)
```

```python
store = denr_funds[['UACS_DPT_DSC','UACS_AGY_DSC','DSC','UACS_OPER_DSC','UACS_FUNDSUBCAT_DSC','UACS_EXP_DSC','UACS_SOBJ_DSC','AMT']]
store.columns = ['Department','Agency','Program','Field Office','Funding Source','Expenses','Specific Item','Amount']
store['Amount'] = store.loc[:,'Amount']*1000
store.to_csv("data/denr.csv",sep="|",index=False)
```

```python
denr_funds['AMT'].sum()*1000
```

```python
comp = denr_funds.groupby("UACS_EXP_DSC").agg({"AMT":"sum"}).reset_index()
comp.columns = ['Components','Amount']
comp.sort_values("Amount",ascending=False)*1000
```

```python
denr_funds.head(2)
```

```python
agency = denr_funds.groupby("UACS_AGY_DSC").agg({"AMT":"sum"}).reset_index()
agency.columns = ['Agency','Amount']
agency.sort_values("Amount",ascending=False)*1000
```

```python
programs = denr_funds.groupby("DSC").agg({"AMT":"sum"}).reset_index()
programs.columns = ['Program','Amount']
programs.sort_values("Amount",ascending=False)*1000
```

```python
funding = programs = denr_funds.groupby("UACS_FUNDSUBCAT_DSC").agg({"AMT":"sum"}).reset_index()
funding.columns = ['Funding Source','Amount']
funding.sort_values("Amount",ascending=False)*1000
```

```python
oper_ofc = denr_funds[["UACS_OPER_DSC","AMT"]].groupby("UACS_OPER_DSC").agg({"AMT":"sum"}).reset_index()
oper_ofc.columns = ['Funding Source','Amount']
oper_ofc.sort_values("Amount",ascending=False)*1000
```

```python
exp_ofc = denr_funds[["UACS_EXP_DSC","AMT"]].groupby("UACS_EXP_DSC").agg({"AMT":"sum"}).reset_index()
exp_ofc.columns = ['Funding Source','Amount']
exp_ofc.sort_values("Amount",ascending=False)*1000
```

```python
obj_ofc = denr_funds[["UACS_EXP_DSC","UACS_SOBJ_DSC","AMT"]].groupby(["UACS_EXP_DSC","UACS_SOBJ_DSC"]).agg({"AMT":"sum"}).reset_index()
obj_ofc.columns = ["Component",'Objective','Amount']
obj_ofc.sort_values("Amount",ascending=False)*1000
```
