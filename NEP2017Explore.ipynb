```python
from openpyxl import *
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.style.use('ggplot')
from pyspark.sql.functions import *
```

```python
#http://www.dbm.gov.ph/wp-content/uploads/NEP2017/2017-NEP.xlsx
files = pd.read_excel("data/2017-NEP.xlsx")
```

```python
data = files[(pd.notnull(files.AMT)==True) & pd.notnull(data.UACS_DPT_DSC)==True]
```

```python
data.to_csv("data/nep2017.csv",sep="|",index=False, encoding="utf-8")
```

```python
budget_per_department = data[['UACS_DPT_DSC','AMT']].groupby(["UACS_DPT_DSC"])['AMT'].agg([np.sum]).reset_index()
budget_per_department.columns = ['Department','Amount']
```

```python
budget_per_department.to_csv("data/per_department.csv",sep="|",index=False)
```
